### Análisis de las vegetaciones afectadas
A continuación se visualiza de forma gráfica la vegetación afectada de los incendios forestales por semestre. Se ve en subgraficos la cantidad de pinos de 0 a 10 años, pinos de 11 a 17 años, pinos de 18 años o más, eucalipto, otras plantaciones, arbolado, matorral, pastizal, vegetación agrícola y desechos afectados por el incendio forestal analizados por meses.

#### Agrupamiento por año y después mes del año correspondiente

```python
dfAgrupado = df.groupby([pd.DatetimeIndex(df["Fecha_Inicio"]).year, pd.DatetimeIndex(df["Fecha_Inicio"]).month]).agg({
    "Pino_0_10": "sum",
    "Pino_11_17": "sum",
    "Pino_18": "sum",
    "Eucalipto": "sum",
    "Otras_Plantaciones": "sum",
    "Arbolado": "sum",
    "Matorral": "sum",
    "Pastizal": "sum",
    "Agricola": "sum",
    "Desechos": "sum"
})
print(dfAgrupado.index)
dfAgrupado
```

```python
# Se crea una lista con la relacion años meses
diccionario_meses = {
    "1" : "ENERO",
    "2" : "FEBRERO",
    "3" : "MARZO",
    "4" : "ABRIL",
    "5" : "MAYO",
    "6" : "JUNIO",
    "7" : "JULIO",
    "8" : "AGOSTO",
    "9" : "SEPTIEMBRE",
    "10" : "OCTUBRE",
    "11" : "NOVIEMBRE",
    "12" : "DICIEMBRE"
}
array_fecha = []
for i, value in enumerate(dfAgrupado.index):
    array_fecha.append(f"{value[0]}-{diccionario_meses[str(value[1])]}")
print(array_fecha)
```

### Comparación de hectáreas quemadas de árboles desde 2017 a 2018

```python
time_cols = ["Pino 0-10 años", "Pino 11-17 años", "Pino 18-n años", "Eucalipto", "Otras plantaciones"]
col_names = ["Pino_0_10", "Pino_11_17", "Pino_18", "Eucalipto", "Otras_Plantaciones"]
colors = ["#f44336", "#673AB7", "#4CAF50", "#FFC107", "#795548"]
leyend = ["Pino de 0 a 10 años", "Pino de 11 a 17 años", "Pino de 18 años o más", "Eucalipto", "Otras plantaciones"]

fig, axs = plt.subplots(len(time_cols), 1, sharex=True, figsize=(13,18))
# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)    

for i, col in enumerate(time_cols):
    axs[i].plot(array_fecha, dfAgrupado[col_names[i]], c = colors[i], marker=".", linewidth=4.0, alpha = 0.6, label = leyend[i])
    axs[i].legend(loc='best')
    axs[i].set_xlabel('Año - Mes', fontsize = 14)
    axs[i].set_ylabel(time_cols[i], fontsize = 13)
    axs[i].grid()
    for index,data in enumerate(dfAgrupado[col_names[i]]):
        value = "%.2f" % data
        axs[i].text(x = index, y = data, s = f"{value}", fontdict=dict(fontsize = 10))
    
    if(i == 0):
        axs[i].set_title("Comparativa de metros cuadrados quemados de árboles por mes de 2017 a 2018",  fontsize = 20)
        
plt.xticks(rotation = 70)
plt.show()
```

### Comparación de vegetación quemada entre 2017 y 2018

```python
time_cols = ["Arbolado", "Matorral", "Pastizal", "Agricola", "Desechos"]
colors = ["#E91E63", "#2196F3", "#009688", "#CDDC39", "#FF9800"]

fig, axs = plt.subplots(len(time_cols), 1, sharex=True, figsize=(13,18))
# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)    

for i, col in enumerate(time_cols):
    axs[i].plot(array_fecha, dfAgrupado[time_cols[i]], c = colors[i], marker=".", linewidth=4.0, alpha = 0.6, label = time_cols[i])
    axs[i].legend(loc='best')
    axs[i].set_xlabel('Año - Mes', fontsize = 14)
    axs[i].set_ylabel(time_cols[i], fontsize = 13)
    axs[i].grid()
    for index,data in enumerate(dfAgrupado[time_cols[i]]):
        value = "%.2f" % data
        axs[i].text(x = index, y = data, s = f"{value}", fontdict=dict(fontsize = 10))
    
    if(i == 0):
        axs[i].set_title("Comparativa de metros cuadrados quemados de vegetación por mes de 2017 a 2018",  fontsize = 20)
        
plt.xticks(rotation = 70)
plt.show()
```

#### Observaciones
En base a los dos gráficos anteriores, se puede apreciar cómo se que se genera un ascenso en la cantidad de superficie quemada a contar del mes de noviembre, para llegar a un peak dado entre los meses de enero y febrero del diferente tipo de vegetación quemada. ### Estudio de los tiempos de duración de incendios forestales

```python
df[["Fecha_Inicio", "Fecha_Extinsion"]].head()
df_copia = df[["Fecha_Inicio", "Fecha_Extinsion", "Region"]]
df_copia
print(df_copia.dtypes)
```

```python
def diff_date(fila):
    if(str(fila["Fecha_Extinsion"]) == "NaT" or str(fila["Fecha_Inicio"]) == "NaT"):
        return np.nan
    else:
        diff = fila["Fecha_Extinsion"] - fila["Fecha_Inicio"]
        dias = int(diff.days)
        horas = int(diff.seconds)//3600
        minutos = (int(diff.seconds)%3600)//60
        segundos = int(diff.seconds) % 60

        if(dias > 0):
            if(dias == 1):
                return int(1)
            elif(dias == 2):
                return int(2)
            else:# Duracion de 3 dias o más
                return int(3)
        else:
            if(segundos >= 0 and horas < 6 ):# Entre 0 y 6 hrs
                return int(-1)
            elif(horas >= 6 and horas < 12):
                return int(-2)
            elif(horas >= 12 and horas < 18):
                return int(-3)
            else:# Duracion entre 18 y 23:59:59 hrs
                return int(-4)

df_copia["Duracion"] = df_copia.apply(diff_date, axis=1)
print(df_copia.dtypes)
print("Cantidad de nulos en columna: ", df_copia["Duracion"].isna().sum())
df_copia.head()
```

```python
# Agrupamiento por rango de horas
dfAgrupado_copia = df.groupby(df_copia["Duracion"]).agg({
    "Region" : "count"
})
dfAgrupado_copia
```

```python
etiquetas = ["Entre 18-24 hrs", "Entre 12-18 hrs", "Entre 6 y 12 hrs", "6 hrs o menos", "1 día", "2 días", "3 días o más"]
color = ["#9C27B0", "#673AB7", "#f44336", "#009688", "#8BC34A", "#FFC107", "#795548"]
explode = [0, 0, 0, 0.1, 0, 0, 0.2]
plt.figure(figsize=(18,10))
plt.pie(dfAgrupado_copia["Region"], labels= etiquetas, autopct="%.2f", explode = explode, colors = color)
plt.title("% de duración de incendios forestales ocurridos entre 2017 - 2018", fontsize = 20)
plt.legend(loc = "best")
plt.show()
```

#### Observaciones
Se puede apreciar como se que la mayoría de los incendios son de dureciones "cortas" (menos de 1 día); pero igual se aprecia como es que la siguiente mayoría está en incendios foretales que duran 3 días o más.

La vegetación más afectada fue la de los pastizales seguida por los pinos de 0 a 10 años y desechos, en otras palabras, vegetación de “baja” altura.

Después se agruparon los datos por semanas. Lo que se agrupó fue la cantidad de incendios ocurridos, el promedio de superficie quemada, porcentaje de humedad, porcentaje de pendiente, velocidad del viento y temperatura. El fin de esta agrupación era poder identificar con mayor precisión el rango de tiempo en que ocurren mayor número de incendios forestales y también para evaluar una posible relación entre la superficie quemada promedio y las otras variables. 

## Número de incendios forestales


La mayor concentración de incendios forestales está entre los meses de noviembre de 2017 y marzo de 2018, en otras palabras, lo que respecta al período de verano. En base a esto se puede apreciar como es que las altas temperaturas son un importante factor en la generación de incendios forestales y también la posible presencia de turistas en dichas zonas geográficas, que producto de algún descuido o mal uso de algún artículo, generen un incendio forestal. 

Después se comparó la superficie quemada con las variables de temperatura, velocidad del viento, porcentaje de pendiente y porcentaje de humedad para evaluar una posible relación entre estas variables.

## Superficie respecto a Temperatura

Se puede apreciar una pequeña relación, respecto a que, a mayor superficie quemada también se dio una alta temperatura. Esta característica se puede apreciar con más claridad en los datos que están en la segunda semana de octubre de 2017 y en los últimos datos. La principal anomalía de esta relación se da en un peak de superficie quemada en el mes de septiembre, donde esta regla tiende a no cumplirse del todo.

## Superficie respecto a Velocidad del viento

Existe una relación más clara que con la temperatura, en el sentido que a mayor superficie quemada, se dio una mayor velocidad del viento, esto se puede apreciar con mucha claridad en los registros que van desde la segunda semana de octubre en adelante. Al igual que con la temperatura, se observa que existe una anomalía en esta relación en el mismo peak de superficie quemada.

## Superficie respecto a Porcentaje de Pendiente

Al igual que con la velocidad, se da una buena afinidad respecto a que un aumento en la superficie quemada, también hubo un mayor porcentaje de pendiente asociada, esto se aprecia a lo largo de todo el gráfico. Particularmente, en este gráfico en el peak de superficie quemada también se da un peak de porcentaje de pendiente de la superficie.

## Superficie respecto al Porcentaje de Humedad

Finalmente, en este gráfico se aprecia una clara correspondencia en que al haber un aumento o descenso de la superficie quemada, también hubo un aumento o descenso en el porcentaje de humedad respectivamente. Esto se aprecia de forma muy evidente en el mínimo que se da para la última semana de agosto y para el peak de la primera semana de septiembre de 2017.

## Duración de los incendios forestales
En relación a la duración de los incendios, se aprecia que la mayoría tiende a durar menos de 6 horas; pero genera una alta preocupación que la 2° mayoría sea de incendios que duraron 3 días o más.

### Análisis geoespacial de los incendios de la Región de Bío-Bío
Al ser la región que concentra el mayor número de incendios, el mapa de calor revela varias zonas con altas concentraciones de incendios forestales. Destacan las ciudades de: Coelemu, Chillán, Tomé, Talcahuano, Concepción, Chiguayante, Coronel, Lota, Yumbel, Laja, Arauco, Curanilahue, Nacimiento, Los Ángeles, Mulchén, Lebu, Cañete y el sur de Cañete
Esta zonas de calor se revelan con mayor claridad al observar el mapa de clustering, donde se revela que cerca de la ciudad de Concepción ocurrieron más de 300 incendios foretales, cerca de la ciudad de Los Ángeles más de 200 incendios forestales y cerca de Curanilahue poco más de 500 incendios forestales.

Por todo lo anteriormente expuesto, primero se puede concluir que existe una mayor relación de dependencia que requiere de un análisis más profundo entre la superficie y la velocidad del viento, porcentaje de pendiente y porcentaje de humedad. Estudiar de forma más detallada estas relaciones, puede resultar de extrema utilidad si se desea crear un modelo de análisis y/o predicción de incendios forestales, en otras palabras, que este modelo se simule en función de las variables de velocidad del viento y porcentaje de pendiente y humedad la dirección más probable que pueda llegar a tomar el fuego en un incendio forestal.
En lo que respecta al análisis geoespacial, los mapas revelan que las ciudades de Quilpué, San Antonio, Cartagena y Valparaíso concentran la mayoría de los incendios forestales de la Región de Valparaíso, por lo que sería prudente mantener sistemas de alerta temprana y monitorio de incendios forestales. Las ciudades de Concepción, Los Ángeles y Curanilahue concentran la mayor cantidad de incendios forestales de la Región del Bío-Bío, por lo que también sería de mucha utilidad sistemas de monitorio y alerta temprana de incendios forestales, sobre todo en las cercanías de Curanilahue, que es en donde hubo la mayor concentración de incendios forestales.
